# Creating a Ordinary Least Squares Regression Model Using Mr. Beast Daily Youtube Stats with 2D Interactive View using Plotly-Dash

```python
# Importing the Libraries

import pandas as pd
import matplotlib.pyplot as plt
import plotly_express as px
```

```python
# Reading the CSV File

df = pd.read_csv('/kaggle/input/mr-beast-youtube-video-statistics/MrBeast_youtube_stats.csv')
```

```python
# Top 10 Rows

df.head(10)
```

```python
'''Using the Plotly_Express Scatter Function to create an OLS, 
which draws an Ordinary Least Squares regression line 
for each discrete-color/symbol group'''

px.scatter(
    data_frame=df,
    x="viewCount",
    y="likeCount",
    color="commentCount",
    height=800,
    width=800,
    hover_data=["title"],
    title="Mr Beast Videos, Likes vs Views",
    trendline= 'ols'
)
```
